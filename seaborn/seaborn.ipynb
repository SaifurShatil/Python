#### Visualizing Statistical Relationships

Understanding relationships between varibales of a dataset. 
 relplot() -> relationship plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a= sns.load_dataset("flights") #from seaborn github
```

```python
sns.relplot(x="passengers",y='month',data=a) # 2d
```

```python
sns.relplot(x="passengers",y="month",hue="year",data=a) # 3d
```

```python
sns.relplot(x="passengers",y="month",hue="year",kind="line",data=a)
```

```python
b= sns.load_dataset("tips")
sns.relplot(x="time",y="tip",data=b, kind="line")
```

#### Visualizing the distribution of a dataset

Understanding dataset with context to be univariate or bivariate.
1. univariate: Univariate is a term commonly used in statistics to describe a type of data which consists of observations on only a single characteristic or attribute. A simple example of univariate data would be the salaries of workers in industry.

2. bivariate data is data on each of two variables, where each value of one of the variables is paired with a value of the other variable. If the variables are quantitative, the pairs of values of these two variables are often represented as individual points in a plane using a scatter plot.

3. displot() -> distribution plot

```python
from scipy import stats
```

```python
c = np.random.normal(loc=5, size=100, scale=2)
```

```python
sns.distplot(c)
```

```python
sns.displot(c)
```

```python
sns.histplot(c)
```

#### Multiplot Grids

To show comparison using same scale and axes
1. FacetGrid()
2. PairGrid()-> pair of variables to compare

```python
a= sns.load_dataset("iris")
b=sns.FacetGrid(a,col="species")
b.map(plt.hist,"sepal_length")
```

```python
a= sns.load_dataset("flights")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

#### Aesthetics
