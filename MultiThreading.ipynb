```python
from threading import *
from time import sleep
```

```python
class Hello():
    def run(self): #name must be run..as threading has run function
        for i in range(5):
            print("Hello")
class Hi():
    def run(self):
        for i in range(5):
            print("Hi")
```

```python
t1=Hello()
t2=Hi()
```

```python
t1.run()
t2.run()
```

```python
class Hello(Thread):
    def run(self):
        for i in range(8):
            print("Hello")
            sleep(1)
class Hi(Thread):
    def run(self):
        for i in range(8):
            print("Hi")
            sleep(1)
```

```python
t1=Hello()
t2=Hi()
t1.start() #t1 thread
sleep(0.2) 
t2.start()  #t2 thread
print('bye') #main thread will execute this
```

```python
t1=Hello()
t2=Hi()

t1.start() #t1 thread
sleep(0.2) 
t2.start()  #t2 thread

t1.join() #to halt main thread
t1.join()

print('bye') #main thread will execute this
```
