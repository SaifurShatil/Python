```python
x1=[3,6,9,12,15]
y1=[10,22,39,41,49]
x2=[3,6,9,12,15]
y2=[7,19,31,44,52]

plt.plot(x1,y1)
plt.plot(x2,y2)

plt.xlabel('x')
plt.ylabel('y')
plt.title("Two Plot")
plt.show()
```

```python
x1=[3,6,9,12,15]
y1=[10,22,39,41,49]
x2=[3,6,9,12,15]
y2=[7,19,31,44,52]

plt.plot(x1,y1, label='First',linewidth=3)
plt.plot(x2,y2, label='Second')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title("Two Plot")
plt.show()
```

```python
plt.scatter(x1,y1,label='F')
plt.scatter(x2,y2, label='S')
plt.legend()
plt.show()
```

```python
plt.hist(x1)
plt.show()
```

```python
x1=[3,6,9,12,15]
y1=[10,22,39,41,49]
x2=[3,6,9,12,15]
y2=[7,19,31,44,52]

plt.plot(x1,y1, label='First',linewidth=3)
plt.plot(x2,y2, label='Second')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0,17)
plt.grid(True)
plt.title("Two Plot")
plt.show()
```

```python
# color
x=[10,20,30]
y=[23,54,78]
plt.plot(x,y,color='r')
plt.show()
```
